### Section 1: The Geometry and Algebra of Vectors

**Def:** A vector is a directed line segment that corresponds to a displacement from one point A to another point B.

The vector from point A to point B is denoted $\vec{AB}$; the point $A$ is called the intial point or tail, and the point $B$ is called the terminal point or head. For simplicity's sake we refer to vectors by their terminal points. By this logic, the set of all points on a plane corresponds to the set of all vectors whose initial point are at the origin. Inversely, this of a point as a vector from the origin to the point.

```python
# import dependencies
import numpy as np
import matplotlib.pyplot as plt

def plotVectors(vecs, cols=['#1190FF'], alpha=1):
    """
    Plot set of vectors.

    Parameters
    ----------
    vecs : array-like
        Coordinates of the vectors to plot. Each vectors is in an array. For
        instance: [[1, 3], [2, 2]] can be used to plot 2 vectors.
    cols : array-like
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
    alpha : float
        Opacity of vectors

    Returns:

    fig : instance of matplotlib.figure.Figure
        The figure of the vectors
    """
    plt.figure()
    plt.axvline(x=0, color='#A9A9A9', zorder=0)
    plt.axhline(y=0, color='#A9A9A9', zorder=0)

    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                   alpha=alpha)
```

```python
v = np.array([[3], [1]])
v

# use the function to plot the vector v
plotVectors([v.flatten()])
plt.ylim(-1, 4)
plt.xlim(-1, 4)
```

It is natural to represent a vector using numbers, such as the coordinates of the terminal point. You can talk about a vector as a column vector, so a vector $v$ with a terminal point at (3, 1) can be denoted, 

$$ \vec{v} = (3, 1) =
  \left[ {\begin{array}{cc}
    3 \\
    1 \\
  \end{array} } \right] $$
  
The individual values in the vector are called components of the vector, just some terminology. The individual coordinates 3 and 1 are the components of the vector $\vec{v}$ in this case.

**Def:** Two vectors are considered equal if and only if their corresponding components are the same. So $[x, y] = [1, 5]$ implies $x = 1$ and $y = 5$. Frequently we refer to vectors as column vectors. There is also the zero vector which is the vector that starts at the origin and ends at the origin and is denoted $\vec{0}$. We say that two vectors are equal if they have the same length and direction. Vectors not on the origin can be equal to vectors elsewhere as long as they are equal in direction and magnitude. Moving forward we're gonna stick with the vectors in standard position, that is the initial point of the vector will be on the origin. Bracket notation or column vec notation assumes the vector is in standard position.

**Def:** The set of all vectors with all components on the $x, y$ plane is denoted $\mathbb{R}^2$.

Now that we know what vectors are, lets talk about the things you can do with vectors.

1. You can add vectors.

Consider $\vec{u} = [u_1, u_2]$ and a $\vec{v} = [v_1, v_2]$, then,

$$ \vec{u} + \vec{v} = [u_1 + v_1, u_2 + v_2] $$

2. You can scale vectors. If $C$ is a scalar, and $\vec{u} = [u_1, u_2]$, then $Cu = [C u_1, C u_2]$. A positive scalar $C$ will retain the direction of $u$ when you mutliply $C$ by $u$. $C = 2$ will double the length of $u$, and $C = \frac{1}{2}$ will shorten $\vec{u}$ by half. A negative $C$ will do the same thing in the opposite direction.

3. You can subtract vectors. This is the same as $\vec{u} + (-1 \times \vec{v})$.



```python
u = np.array([[3], [-1]]) # orange
v = np.array([[1], [4]]) # blue
print(u + v) # green

# use the function to plot the vector v
plotVectors([v.flatten(), u.flatten(), (u + v).flatten()], cols=['#1190FF', '#FF9A13', '#00ff00'])
plt.ylim(-1, 5)
plt.xlim(-1, 5)
```

```python
# multiplying a vector by a scalar
v = np.array([[1], [2]])

plotVectors([v.flatten(), 2 * v.flatten()], cols=['#1190FF', '#FF9A13'])
plt.ylim(-1, 5)
plt.xlim(-1, 5)
```

```python
# subtracting vectors, u - v
u = np.array([[3], [-1]]) # orange
v = np.array([[1], [4]]) # blue
print(u + v) # green

# use the function to plot the vector v
plotVectors([v.flatten(), u.flatten(), (u - v).flatten()], cols=['#1190FF', '#FF9A13', '#00ff00'])
plt.ylim(-5, 5)
plt.xlim(-1, 5)
```

Everything we've done easily generalizes to $\mathbb{R}^3$. The set of all ordered tiples of real numbers is denoted $mathbb{R}^3$. Vectors in $mathbb{R}^3$ are given 3 components/coordinates, 

$$ \vec{v} = [x, y, z] $$

We can talk about systems of any number of equations $m$, and any number of unkowns $n$. Now we are in $n$ dimensional space, $\mathbb{R}^n$. In general, we define $\mathbb{R}^n$ as the set of all ordered real numbers written as row or column vectors. Thus $v$ in $\mathbb{R}^n$ takes on the form, 

$$ \vec{v} =
  \left[ {\begin{array}{cc}
    v_1 \\
    v_2 \\
    \vdots \\
    v_n
  \end{array} } \right] $$
  
We retain addition and multiplication in this $\mathbb{R}^n$ space.

**Theorem:** Consider three vectors $\vec{u}, \vec{v}, \vec{w} \in \mathbb{R}^n$, and let $c, d$ be scalars. Then,

- $u + v = v + u$
- $(u + v) + w = u + (v + w)$
- $u + (-u) = 0$
- $c(u + v) = cu + cv$
- $(c + d)u = cu + du$
- $(cd)u = c(du)$
- $1 \times u = u$

**Def:** Linear Combination

A linear combination of some vectors $\vec{u}, \vec{v}, \vec{w}$ is the sum of some combination of the vectors, like 2 $\vec{v}$ and 4 of $\vec{u}$ and only 1 of $\vec{w}$. You know this stuff.

### Section 1.2: Length and angle, and the dot product.

Consider,

$$ \vec{u} =
  \left[ {\begin{array}{cc}
    u_1 \\
    u_2 \\
    \vdots \\
    u_n
  \end{array} } \right], 
\vec{v} =
  \left[ {\begin{array}{cc}
    v_1 \\
    v_2 \\
    \vdots \\
    v_n
  \end{array} } \right] $$

The dot product of $\vec{u}$ and $\vec{v}$ is,

$$ \vec{u} \cdot \vec{v} = u_1v_1 + y_2v_2 + \dots + u_nv_n $$

**Theorem:** Let $\vec{u}, \vec{v}, \vec{w} \in \mathbb{R}^n$, and $c$ is a scalar,

- $u \cdot v = v \cdot u$
- $u \cdot (v + w) = u \cdot v + u \cdot w$
- $(cu) \cdot v = c(u \cdot v) = u \cdot cv$
- $u \cdot u \geq 0$ and $u \cdot u = 0$ iff $u = \vec{0}$

The length of a vector $\vec{v} = [a, b]$ is defined as $\sqrt{a^2 + b^2}$ by the pythagorean theorem. Don't forget that the pythagorean theorem works in more than 2 dimensions. You can also take the square root of the dot product of a vector and itself to get its length!

$$ ||\vec{v}|| = \sqrt{\vec{v} \cdot \vec{v}} = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2} $$

**Def:** A vector whose length is one is called a unit vector. Given any vector $\vec{v}$, we can divide by its length to get the unit vector while retaining the same direction. This process is called normalizing the vector $\vec{v}$.

**Cauchy-Shwarz Inequality:** If $u, v \in \mathbb{R}^n$ then $|u \cdot v| \leq ||u|| ||v||$.

**The Triangle Inequality:** The sum of any two sides of a triange is always greater than or equal to the third side of the triangle. In other words, $||u|| + ||v|| \geq ||u + v||$.

Recall the distance between two points $(a_1, b_1)$ and $(a_2, b_2)$ is $\sqrt{(a_2 - a_1)^2 + (b_2 - b_1)^2}$ We'll motive distance with a general $\mathbb{R}^n$. In terms of vectors, if you have $u = [a_1, b_1]$ and $v = [a_2, b_2]$. The distance between the terminal points of these vectors is,

$$ u - v = [a_2 - a_1, b_2 - b_1] $$
$$ ||u - v|| = \sqrt{(a_2 - a_1)^2 + (b_2 - b_1)^2} $$

**Def:** If $u, v \in \mathbb{R}^n$, the distance between these two vectors $u, v$ is $d(u, v) = ||u - v||$.  

Example:
Find the distance between

$$ \vec{u} =
  \left[ {\begin{array}{cc}
    \sqrt{2} \\
    1 \\
    -1
  \end{array} } \right], \vec{v} = \left[ {\begin{array}{cc}
    0 \\
    2 \\
    -2
  \end{array} } \right] $$
  
  
$$ ||u - v|| = \left[ {\begin{array}{cc}
    \sqrt{2} \\
    -1 \\
    1
  \end{array} } \right] = \sqrt{(\sqrt{2})^2 + (-1)^2 + 1^2} = 2 $$
  
**Angles** Let's consider $\mathbb{R}^2$. If $u, v$ are non-zero vectors, we will refer to the angle between $u$ and $v$ as the agle $\theta$ determined by these vectors that satisfies $0 \leq \theta \leq 180$.

The **Law of Cosines** considers the length of the sides of a triangle to determine the angle between the two vectors.

$$ ||u-v||^2 = ||u||^2 + ||v||^2 - 2(||u||) (||v||) cos(\theta) $$

Last time we said that the length of a vector squared is the dot product of a vector with itself. Lets rewrite this equation as a dot product.

$$ (u - v) \cdot (u - v) = u \cdot u + v \cdot v - 2 (||u||) (||v||) cos(\theta) $$

Simplifying,

$$ u \cdot v = ||u|| ||v|| cos(\theta) $$

Taking the inverse cosine,

$$ cos(\theta) = \frac{u \cdot v}{||u|| ||v||}  $$

This is the definition of the angle between two vectors in arbitrary n-dimensional space. There are two angle between two vectors, we want the one $0 \leq \theta \leq 180$.

Example: Let's compute the angle between $u = [2, 1, -2]$ and $v = [1, 1, 1].

$$ cos(\theta) = \frac{(2)(1) + (1)(1) + (-2)(1)}{3\sqrt{3}} $$
$$ \theta = cos^{-1} (\frac{1}{3 \sqrt{3}}) $$

Two non-zero vectors $u$ and $v$ are perpendicular, or orthogonal, if the angle between them is 90 degrees, or $\frac{\pi}{2}$. Thus for perpendicular vectors,

$$ \frac{u \cdot v}{||u|| ||v||} = cos(\frac{\pi}{2}) = 0 $$

**Def:** Two vectors $u$ and $v$ in $\mathbb{R}^n$ are orthogonal when the dot product is zero, you have perpendicular vectors. 

We can get an equivalent definition for the dot product, $ u \cdot v = (||u||)(||v||) cos(\theta)$. 

## Projection

We consider the problem of finding the distance from a point to a line. You have a point B and a line L. You would like to know the shortest distance between the point and the line, that is, the perpendicular distance. Lets call this point p. We want to find the length of of this vector perpendicular to L and with terminal points b and p, or $||\vec{bp}||$. 

The formula for the projection in $\mathbb{R}^2$ is,

$$ p = \frac{u \cdot v}{u \cdot u} u$$

**Def:** If $u. v \in \mathbb{R}^n$ and $u \neq 0$, then the projection of $v$ onto $u$, denoted $proj_u v$, is

$$ proj_u v = \frac{u \cdot v}{u \cdot u} u $$

### Section 1.3: Lines and Planes

We know what a line and a plane is, but we're gonna frame them as vectors and systems of equations. Let's discuss lines in $\mathbb{R}^2$ and $\mathbb{R}^3$. We're familiar with $y = mx + b$, but with linear algebra be familiar with the general form of a line, $ax + by = 0$. 

Example:
Consider the line $2x + y = 0$. The left hand side looks like a dot product. 

$$ 2x + y = n \cdot x = \left[ {\begin{array}{cc}
    2 \\
    1 \\
  \end{array} } \right] \cdot \left[ {\begin{array}{cc}
    x \\
    y \\
  \end{array} } \right] = 0 $$'
  
The dot product here equals 0. The line you are after is the set of all vectors that are perpendicular to $n = \left[ {\begin{array}{cc}
    2 \\
    1 \\
  \end{array} } \right]$! All the vectors that are perpendicular to $n$ are on this line in (standard form, initial point on the origina). The equation $n \cdot x$, where $x = \left[ {\begin{array}{cc}
    x \\
    y
  \end{array} } \right]$, is called the *normal form* of the equation of the line. Notice how all the vectors on this line are some scalar multiple of a single vector. We are gonna call this vector a direction vector. Namely, in this setup, if the vector $d = \left[ {\begin{array}{cc}
    1 \\
    -2
  \end{array} } \right]$ then all other vectors on this line $2x + y = 0$ are multiples to the direction vector $d$. Any vector on the line can be defined by $X = \left[ {\begin{array}{cc}
    x \\
    y
  \end{array} } \right] = td$ where $t$ is any scalar. The term $td$ is called the vector form of a line. 
  
What about lines that do not go through the origin? Consider the line $2x + y = 5$. We can still find a line perpendicular to this line, but this time the vectors that define the points on the line (initial point on the origin) will not be perpendicular to $n$. This detail matters. Below is a general definition.

**Def:** The *normal form* of the equation of a line $l \in \mathbb{R}^2$ is $n \cdot (x - p) = 0$. Foiling, you get $n \cdot x = n \cdot p$ where $p$ is any fixed point on the line and $n$ is a normal vector on the line. The general form pf the equation of a line is $ax + by = c$ where $[^a_b]$ is a normal vector for the line.

What about the vector form of $ax + y = 5$? We fix a point $P$ on the line then if $X = \left[ {\begin{array}{cc}
    x \\
    y
  \end{array} } \right]$ is on the line, then $X - P$ is a multiple of any direction vector $d$, $X - P = td$ for some scalar $t$. 
  
**Def:** The *vector form* of the equation of the line $X = P + td$ where $P$ is a specified point on the line and $d$ is a direction vector for $l$.

The equations corresponding to the components of $X - P + td$ are called parametric equations of $l$.

The parametric equations are then

**Planes in $\mathbb{R}^3$**

As a priliminary question, what does a line look line in $\mathbb{R}^3$ mathematically. In $\mathbb{R}^2$ the general form is $ax + by = x$. What is it in $\mathbb{R}^3$? This is an intuitive exercise. In $\mathbb{R}^3$ there is a plane perpendicular to a a given point on a line! So lets discuss this. Precisely, every plane $\mathcal{P}$ in $\mathbb{R}^3$ can be determined by specifying a point $p$ on the plane and a non-zero vector $n$ perpendicular to $\mathcal{P}$. So what is the equation of a plane specifically? If $X = \left[ {\begin{array}{cc}
    x \\
    y \\
    z
  \end{array} } \right]$ is a point on the plane $\mathcal{P}$, then $n \cdot (x - p) = 0$ is perpendicular to the normal vector $n$. And that is what it means to be perpendicular to a plane. $n \cdot (x - p) = 0$ becomes,
  
$$ n \cdot x = n \cdot p $$
$$ \left[ {\begin{array}{cc}
    a \\
    b \\
    c
  \end{array} } \right] \cdot \left[ {\begin{array}{cc}
    x \\
    y \\
    z
  \end{array} } \right] = n \cdot p$$
$$ ax + by + cz = n \cdot p $$

*This* is the general form equation of a plane.

Let's just plug it in.

$$ n \cdot (x - p) = 0 $$

    3

We can also express the equation of a plane in vector or parametric form. There are so many directional vectors to describe a plane, but all we need are two directional vectors $u, v$. Let $X = \left[ {\begin{array}{cc}
    x \\
    y \\
    z
  \end{array} } \right]$ be a point on the plane. Any vector on this plane will be a linear combination of the direction vectors $u$ and $v$. We'll prove this later. That is, $X - p = su + tv$.
  
**Def:** The vector form of the equation of a plane $\mathcal{P}$ in $\mathbb{R}^3$ is,

$$ X = p + su + tv $$

Where $X = \left[ {\begin{array}{cc}
    x \\
    y \\
    z
  \end{array} } \right]$, $p$ is a specified point on the $\mathcal{P}$, and $u$ and $v$ are non-parallel direction vectors for $\mathcal{P}$. The equations corresponding to the components of the vector form of the equation are called *parametric equations* of $\mathcal{P}$.

$$ x + 2y + 3z = 9$$

$$ P = (1, 1, 2) $$
$$ Q = (0, 0, 3 $$
$$ R = (9, 0, 0) $$
$$ \vec{u} = \vec{PQ} = Q - P = (-1, -1, 1) $$
$$ \vec{v} = \vec{PR} = R - P = (8, -1, -2) $$ 

We have the equation $X = p + su + tv$.
